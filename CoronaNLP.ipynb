```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer


nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize stop words and lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()


# Function to clean text
def clean_tweet(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text)  # Remove @mentions
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'#[A-Za-z0-9]+', '', text)  # Remove hashtags
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # Remove special characters and numbers
    text = text.lower()  # Lowercase the text
    text = text.split()  # Tokenize the text
    text = [lemmatizer.lemmatize(word) for word in text if word not in stop_words]  # Lemmatize and remove stopwords
    text = ' '.join(text)  # Rejoin the cleaned words
    return text

# Apply the cleaning function to 'OriginalTweet' column
df['CleanedTweet'] = df['OriginalTweet'].apply(clean_tweet)

# Show the cleaned 'OriginalTweet' column
print(df[['OriginalTweet', 'CleanedTweet']].head())
```
